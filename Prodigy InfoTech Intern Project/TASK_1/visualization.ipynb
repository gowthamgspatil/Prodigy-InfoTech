---

##  Examples
1. **Categorical Example (Bar Chart)**  
   - **X-axis** → Gender (Male, Female, Other)  
   - **Y-axis** → Number of people in each category  

2. **Continuous Example (Histogram)**  
   - **X-axis** → Age bins (e.g., 0–10, 11–20, 21–30, etc.)  
   - **Y-axis** → Count of people in each bin

## Importing the necessary python libraries and the data set.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
population_data = pd.read_csv(
    'API_SP.POP.TOTL_DS2_en_csv_v2_38144.csv',
    skiprows=4
)
```

```python
population_data 
```

## Displaying the data set

```python
population_data.head(266)
```

```python
population_data.describe
```

### Extracting the relevent columns

```python
data_2022 = population_data.loc[:, ['Country Name', '2022']]
```

## Remove rows with missing values

```python
data_2022.dropna()
```

## Display the first few rows of the cleaned data

```python
data_2022.head()
```

## We will then visualize the data using a Histogram.

```python
sns.set(style="whitegrid")

plt.figure(figsize = (12, 6))
sns.histplot(data = data_2022, x='2022', bins=30, kde=True, color='blue')
plt.title('Population Distribution in 2022')
plt.xlabel('Population')
plt.ylabel('Frequency')
plt.show()
```

## We will then visualize the data using a Bar Chart.

```python
data_2022['2022'] = pd.to_numeric(data_2022['2022'], errors='coerce')

top_10_countries = data_2022.sort_values(by='2022', ascending=False).head(10)

sns.set(style="whitegrid")

plt.figure(figsize=(12, 6))
sns.barplot(data = top_10_countries, x='Country Name', y='2022', palette='viridis')
plt.title('Top 10 Most Populous Countries in 2022')
plt.xlabel('Country')
plt.ylabel('Population')
plt.xticks(rotation=45)
plt.show()
```
